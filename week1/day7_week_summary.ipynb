```python
pollution_data = {
    "Central": 42,
    "North": 35,
    "East": 50,
    "South": 28,
    "West": 33
}

pollution_data
```

```python
max_value = max(pollution_data.values())


min_value = min(pollution_data.values())


avg_value = sum(pollution_data.values()) / len(pollution_data)

print("Максимум:", max_value)
print("Минимум:", min_value)
print("Среднее:", avg_value)
```

```python
print("Районы со значением выше среднего:")
for district, value in pollution_data.items():
  if value > avg_value:
    print(district, ":" , value)
```

```python
def summary_stats(values):
    return {
        "min": min(values),
        "max": max(values),
        "avg":sum(values)/ len(values) 
    }
    summary_stats(list(pollution_data.values()))
```

```python
print("------- Итоговый отчёт -------")
stats = summary_stats(list(pollution_data.values()))
print("Среднее загрязнение:", stats["avg"])
print("Максимум:", stats["max"])
print("Минимум:", stats["min"])
print()

print("Районы с превышением 40:")
for d, v in pollution_data.items():
    if v > 40:
        print(d, v)
```
