```python
def mean_value(values):
    total = 0 
    for v in values: 
     total += v
    return total / len(values)
    mean_value([10,20,30,40])
```

```python

pollution = [12, 18, 31, 22, 15, 40, 27]

print("Минимум", min(pollution))
print("Максимум", max(pollution))
print("Среднее", mean_value(pollution))
print("Значение выше 20:")
for p in pollution:
    if p > 20: 
        print(p)
```
