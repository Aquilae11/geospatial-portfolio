```python

weather = { 
"monday": 5,
"tuesday": 7,
"wednesday": 3,
"thursday": 6,
"friday": 9,
"saturday": 10,
"sunday": 4,
}
weather
```

```python

max_value = max(weather.values())
for day, temp in weather.items():
    if temp == max_value:
        print("максимальная температура:",temp, "была в день:", day)
```

```python
avg = sum(weather.values()) / len(weather)
print("Средняя температура за неделю:", avg)
```

```python
print("Дни с температурой ниже 6 градусов:")

for day, temp in weather.items():
    if temp < 6:
        print(day, ":", temp)
```
