```python
import pandas as pd

data = {
    "district": ["Central", "North", "East", "South", "Central", "East", "South"],
    "pm25": [42, 35, 50, 28, 40, 52, 33],
    "population": [120000, 90000, 150000, 80000, 125000, 150000, 85000]
}

df = pd.DataFrame(data)
df
```

```python
df.groupby("district")
```

```python
df.groupby("district")["pm25"].mean()
```

```python
df.groupby("district").agg({
    "pm25": ["mean", "max", "min"],
    "population": "mean"
})
```

```python
high_pm25 = df[df["pm25"] > 40]
high_pm25.groupby("district")["pm25"].mean()
```

```python
df["district"].value_counts()
```

```python
df.groupby("district")["pm25"].mean().sort_values(ascending=False)
```

```python
summary = df.groupby("district").agg({
    "pm25": ["mean", "max", "min"],
    "population": "mean"
})

summary.to_csv("district_summary.csv")
summary
```
